```python
simplified_df = df[['title','abstract','group']]
simplified_df.head()
```

```python
# Create a translation table to remove punctuation
translator = str.maketrans('', '', string.punctuation)
simplified_df.loc[:,'title'] = simplified_df['title'].str.translate(translator).str.lower().str.strip()
simplified_df.loc[:,'abstract'] = simplified_df['abstract'].str.translate(translator).str.lower().str.strip()

# Display the first few rows to verify the changes
simplified_df.head()
```

## Stopwords

```python
# Download required NLTK data (run once)
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

nltk.download('punkt_tab')

try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

# Get English stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words from text
def remove_stopwords(text):
    if pd.isna(text):
        return text
    
    # Tokenize the text
    words = word_tokenize(text.lower())
    
    # Remove stop words and join back
    filtered_words = [word for word in words if word.lower() not in stop_words]
    
    return ' '.join(filtered_words)
```

```python
# Apply stop word removal to title and abstract columns
simplified_df['title_no_stopwords'] = simplified_df['title'].apply(remove_stopwords)
simplified_df['abstract_no_stopwords'] = simplified_df['abstract'].apply(remove_stopwords)

# Display the first few rows to verify the changes
simplified_df.head()
```

```python
# Download all required NLTK data
required_packages = ['punkt', 'wordnet', 'averaged_perceptron_tagger', 'omw-1.4']

for package in required_packages:
    try:
        nltk.data.find(f'tokenizers/{package}' if package == 'punkt' else f'taggers/{package}' if 'tagger' in package else f'corpora/{package}')
    except LookupError:
        print(f"Downloading {package}...")
        nltk.download(package)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# Función para mapear POS de nltk al formato de wordnet
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Por defecto, como antes
```

```python
# Función mejorada de lematización con POS
def lemmatize_text_with_pos(text):
    if pd.isna(text):
        return text

    words = word_tokenize(text.lower())
    pos_tags = nltk.pos_tag(words)

    lemmatized_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in pos_tags
    ]
    
    return ' '.join(lemmatized_words)

```

```python
# Apply lemmatization to the columns
simplified_df['title_lemmatized'] = simplified_df['title_no_stopwords'].apply(lemmatize_text_with_pos)
simplified_df['abstract_lemmatized'] = simplified_df['abstract_no_stopwords'].apply(lemmatize_text_with_pos)

# Display the first few rows to verify
simplified_df[['title_no_stopwords', 'title_lemmatized', 'abstract_no_stopwords', 'abstract_lemmatized']].head(20)
```
